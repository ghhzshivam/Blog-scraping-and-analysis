```python
import pandas as pd
from bs4 import BeautifulSoup as bs
import re
import requests
```

```python
# Function for web Scraping

def get_data(link):
    headers = {'User-Agent':'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/92.0.4515.131 Safari/537.36'}
    # open that page:
    rawpage = requests.get(link,headers=headers)

    #make a BS object from the html page
    soup = bs(rawpage.content, 'html5lib')
    
    # Getting title
    title = soup.body.findAll('h1')
    title = [tit.get_text(' ', strip = True) for tit in title]
    
    #  geting body from blog
    body = soup.body.findAll('p')
    content = [bod.get_text(' ', strip = True) for bod in body]
    
    # Creating a string for file 
    title = title+content    
    txt = ' '.join(a for a in title)
    
    return txt    
```

```python
df = pd.read_excel('input.xlsx')
df
```

```python
# Getting the text from online blog and converting it into a file and saving it

count = 1
Error_list = []
for link in df['URL']:   
    try:
        txt = get_data(link)
        
    except:
        with open('data/{0}.txt'.format(count), 'w+',encoding="utf-8")as f:
                f.write('Loss')
        Error_list.append(count)
    else:
        with open('data/{0}.txt'.format(count), 'w+', encoding="utf-8")as f:
                f.write(txt)
    
        count += 1
    finally:
        print(count)
        
```
